8. **Using the `Rocket` class which we created as part of the week 7 class activity, we are going to create a new class called shuttle. The shuttle class should inherit all of Rocket's characteristics but has one additional parameter: flights_completed which measures the number of flights the shuttle has completed.
   Define the shuttle class as the child of the rocket class (please define it in the same jupyter notebook which has all of your other HW answers), and then initialize a new instance of your shuttle class. Once you've created a Shuttle, print the following message: "The shuttle is located at (__, __), and has recorded __ flights", filling the blank spaces to reflect the information for the Shuttle you just created.**

```python
import math
```

```python
class Rocket:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    def move_rocket(self, move_x=0, move_y=0):
        self.x += move_x
        self.y += move_y
    def calc_distance(self, other_rocket):
        distance = math.sqrt((self.x-other_rocket.x)**2 + (self.y-other_rocket.y)**2)
        return distance
```

```python
class Shuttle(Rocket):
    def __init__(self, x=0, y=0, flights_completed=0):
        super().__init__(x, y)
        self.flights_completed = flights_completed
```

```python
shuttle_example = Shuttle(5, 5, 5)
print(f"The shuttle is located at ({shuttle_example.x}, {shuttle_example.y}), and has recorded {shuttle_example.flights_completed} flights")
```

9. **Write a single for loop which randomly creates 11 rockets and 11 shuttles using randomly generated integers. Store this output in two separate lists. Note you can use the randint() function from the random module (which you can import with the command from random import randint) to randomly generate the characteristics of your Rockets and Shuttles.**

```python
from random import randint
```

```python
rocket_list = []
shuttle_list = []
for i in range(11):
    rocket_x = randint(0,50)
    rocket_y = randint(0,50)
    
    shuttle_x = randint(0,50)
    shuttle_y = randint(0,50)
    flights = randint(0,50)

    rocket_list.append(Rocket(rocket_x, rocket_y))
    shuttle_list.append(Shuttle(shuttle_x, shuttle_y, flights))
```

```python
for index, shuttle in enumerate(shuttle_list):
    print(f"Shuttle {index} has completed {shuttle.flights_completed} flights.")
```

```python
for index, shuttle in enumerate(shuttle_list):
    distance = shuttle.calc_distance(shuttle_list[0])
    print(f"The first shuttle is {distance:.2f} units away from shuttle {index}.")
```
