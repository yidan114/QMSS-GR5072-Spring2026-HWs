## Q3. SalePrice grouped by a categorical variable (2-4 categories)

```python
print("Categorical variables with 2-4 unique categories:")
for col in df_cat.columns:
    n_unique = df_cat[col].nunique()
    if 2 <= n_unique <= 4:
        print(f"  {col}: {n_unique} categories -> {df_cat[col].unique()}")
```

```python
# Choose LandContour (flatness of the property) - 4 categories, likely related to SalePrice
# Create SalePrice dataset
df_saleprice = df_num[["SalePrice"]]

# Combine SalePrice with the chosen categorical variable
df_combined = pd.concat([df_saleprice, df[["LandContour"]]], axis=1)

# Mean, Median, SD of SalePrice grouped by LandContour
grouped_stats = df_combined.groupby("LandContour")["SalePrice"].agg(["mean", "median", "std"])
grouped_stats.columns = ["Mean", "Median", "SD"]
print("SalePrice statistics by LandContour:")
display(grouped_stats)

print("\nYes, the average sale price appears to vary by land contour.")
print("Properties that are near flat/level (Lvl) or hillside (HLS) tend to have")
print("different average prices compared to those on banked (Bnk) or low depression (Low) lots.")
```

## Q4. Regress SalePrice on LotArea, OverallQual, OverallCond

```python
import statsmodels.api as sm

# Specify DVs and IVs
x = df_num[['LotArea', 'OverallQual', 'OverallCond']]
y = (df_num["SalePrice"])
x = sm.add_constant(x)

# fit linear regression model
model = sm.OLS(y, x).fit()

# view model summary
print(model.summary())
```

## Q5. For loop: Regression by each level of the categorical variable

```python
# Get unique 
categories = df["LandContour"].unique()
print(f"Categories: {categories}\n")
model_summaries = []

for cat in categories:
    subset = df[df["LandContour"] == cat]
    
    # regression
    x = subset[['LotArea', 'OverallQual', 'OverallCond']]
    y = subset["SalePrice"]
    x = sm.add_constant(x)
    
    # fit model
    model = sm.OLS(y, x).fit()
    model_summaries.append(model.summary())

for i, summary in enumerate(model_summaries):
    print(f"\n{'='*60}")
    print(f"Regression for LandContour = {categories[i]}")
    print(f"{'='*60}")
    print(summary)
    print("\n")
```

**Does the effect of LotArea on SalePrice vary within the different levels of LandContour?**

Yes, the coefficient for LotArea differs across the LandContour groups. To properly compare, we need to consider both the coefficient and its standard error. If the confidence intervals for the LotArea coefficient do not overlap across groups.

**What is another, more common way to answer a question like this?**

A more common approach is to include an interaction term between LotArea and LandContour in a single regression model. This tests whether the relationship between LotArea and SalePrice is moderated by LandContour.

## Q6. For loop: Frequency and relative frequency tables for categorical variables

```python
# For loop
for col in df_cat.columns:
    print(f"Frequency table for {col}===============")
    freq = df_cat[col].value_counts()
    print(freq)
    
    # frequency
    print(f"\nRelative Frequency table for {col}===============")
    rel_freq = df_cat[col].value_counts(normalize=True)
    print(rel_freq)
```

## Q7. While loop: Frequency and relative frequency tables for categorical variables

```python
# While loop
i = 0
cat_columns = list(df_cat.columns)

while i < len(cat_columns):
    col = cat_columns[i]
    
    print(f"Frequency table for {col}===============")
    freq = df_cat[col].value_counts()
    print(freq)
    
    print(f"\nRelative Frequency table for {col}===============")
    rel_freq = df_cat[col].value_counts(normalize=True)
    print(rel_freq)
    
    print("\n\n")
    
    i += 1
```

## Q8. Shuttle class inheriting from Rocket class

```python

class Rocket:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    def move_rocket(self, x_increment=0, y_increment=1):
        self.x += x_increment
        self.y += y_increment
    
    def get_distance(self, other_rocket):
        distance = sqrt((self.x - other_rocket.x)**2 + (self.y - other_rocket.y)**2)
        return distance


# Define shuttle
class Shuttle(Rocket):
    def __init__(self, x=0, y=0, flights_completed=0):
        super().__init__(x, y)
        self.flights_completed = flights_completed


# Init
my_shuttle = Shuttle(x=5, y=10, flights_completed=3)
print(f"The shuttle is located at ({my_shuttle.x}, {my_shuttle.y}), and has recorded {my_shuttle.flights_completed} flights.")
```

## Q9. Randomly create 11 Rockets and 11 Shuttles

```python
from random import randint

rockets = []
shuttles = []

for i in range(11):
    rockets.append(Rocket(x=randint(0, 100), y=randint(0, 100)))
    shuttles.append(Shuttle(x=randint(0, 100), y=randint(0, 100), flights_completed=randint(0, 10)))
```

```python

for i, shuttle in enumerate(shuttles):
    print(f"Shuttle {i} has completed {shuttle.flights_completed} flights.")
```

```python


for i, shuttle in enumerate(shuttles):
    distance = shuttles[0].get_distance(shuttle)
    print(f"The first shuttle is {distance:.2f} units away from shuttle {i}.")
```
