```python
import pandas as pd
import numpy as np

nba = pd.read_csv("../Data/Seasons_Stats_NBA.csv")
nba
```

#### 1. Which NBA player scored the most points in a single season? What seasons/year was this? <a class="anchor" id="bullet1"></a>

```python
q1 = (
    nba.loc[nba["PTS"] == nba["PTS"].max(), ["Player", "Year", "PTS"]]
       .drop_duplicates()
       .sort_values(["Player", "Year"])
       .reset_index(drop=True)
)

display(q1.head())
display(q1.tail())
q1
```

```python
q1["Year"] = q1["Year"].astype(int)
q1
```

#### 2. Which NBA player scored the most points in 1991? <a class="anchor" id="bullet2"></a>

In 1991, Michael Jordan scored the most points.

```python
q2_pool = nba.loc[nba["Year"] == 1991, ["Player", "Year", "PTS"]].copy()
q2_pool = q2_pool.dropna(subset=["PTS"])

q2 = (
    q2_pool.loc[q2_pool["PTS"] == q2_pool["PTS"].max()]
           .drop_duplicates()
           .sort_values(["Player"])
           .reset_index(drop=True)
)

q2["Year"] = q2["Year"].astype(int)

display(q2.head())
display(q2.tail())
q2
```

#### 3. Which player had the best free throw percentage ("FT%") from the year 2000 to the most recent year in the data? <a class="anchor" id="bullet3"></a>

There are many players who have a FT% of 100%, below is a list of these players sorted by their name (Player). Don't need to sort, but it is nice to see who had 100% FT% while shooting the most shots.

```python
most_recent_year = int(nba["Year"].max())
q3_pool = nba.loc[(nba["Year"] >= 2000) & (nba["Year"] <= most_recent_year),
                  ["Player", "Year", "FT%"]].copy()

q3_pool = q3_pool.dropna(subset=["FT%","Player","Year"])

best_ft = q3_pool["FT%"].max()

q3 = (
    q3_pool.loc[q3_pool["FT%"] == best_ft, ["Player","Year","FT%"]]
           .drop_duplicates()
           .sort_values(["Player"]) 
           .reset_index(drop=True)
)

q3["Year"] = q3["Year"].astype(int)

display(q3.head())
display(q3.tail())
q3
```

#### 4. Among those with at least 50 attempts ('FTA'), which player had the best free throw percentage ("FT%") from the year 2000 to the most recent year in the data? <a class="anchor" id="bullet4"></a>

Among those with at least 50 free throw attempts (FTA), Terrell Brandon had the best FT% since 2000.

```python
most_recent_year = int(nba["Year"].max())
q4_pool = nba.loc[(nba["Year"] >= 2000) & (nba["Year"] <= most_recent_year) & (nba["FTA"] >= 50),
                  ["Player", "Year", "FTA", "FT%"]].copy()

q4_pool = q4_pool.dropna(subset=["FT%","Player","Year","FTA"])

best_ft_50 = q4_pool["FT%"].max()

q4 = (
    q4_pool.loc[q4_pool["FT%"] == best_ft_50, ["Player","Year","FTA","FT%"]]
           .drop_duplicates()
           .sort_values(["Player"])
           .reset_index(drop=True)
)

q4["Year"] = q4["Year"].astype(int)

display(q4.head())
display(q4.tail())
q4
```

#### 5. Rename the variable "Pos" to "Position". 

```python
nba = nba.rename(columns={"Pos": "Position"})

display(nba[["Position"]].head())
display(nba[["Position"]].tail())
```

#### 6. Use this variable to create two variables that are called "first_position" and "second_position".

```python
pos_split = nba["Position"].astype("string").str.split("-", n=1, expand=True)
nba["first_position"] = pos_split[0]
nba["second_position"] = pos_split[1] 

display(nba[["Position", "first_position", "second_position"]].head())
display(nba[["Position", "first_position", "second_position"]].tail())
```

#### 7. Combine these two variables back into a single variable called "position_united".

```python
nba["position_united"] = np.where(
    nba["second_position"].isna() | (nba["second_position"].astype("string").str.len() == 0),
    nba["first_position"].astype("string"),
    nba["first_position"].astype("string") + "-" + nba["second_position"].astype("string")
)

pos_freq = (
    nba["position_united"]
      .value_counts(dropna=False)
      .rename_axis("position_united")
      .reset_index(name="count")
)

display(pos_freq.head())
display(pos_freq.tail())
pos_freq
```

#### 8. Create two new datasets. <a class="anchor" id="bullet8"></a>

```python
n = len(nba)
mergeid = pd.Series(range(1, n + 1), name="mergeid")

nba_no_age = nba.drop(columns=["Age"]).copy()
nba_no_age.insert(0, "mergeid", mergeid)  # put mergeid as first column

nba_year_player_age = nba[["Year", "Player", "Age"]].copy()
nba_year_player_age.insert(0, "mergeid", mergeid)

display(nba_no_age.head())
display(nba_no_age.tail())

display(nba_year_player_age.head())
display(nba_year_player_age.tail())
```

#### 9. Join the two datasets from question (8) together to recreate the original dataset plus the new merge id. <a class="anchor" id="bullet9"></a>

```python
nba_recreated = (
    nba_no_age.merge(nba_year_player_age, on="mergeid", how="inner")
)

print("original rows:", nba.shape[0], "recreated rows:", nba_recreated.shape[0])
print("has mergeid:", "mergeid" in nba_recreated.columns)

display(nba_recreated.head())
display(nba_recreated.tail())
```

```python
nba_recreated_clean = nba_recreated.copy()

nba_recreated_clean = nba_recreated_clean.drop(columns=["Year_y", "Player_y"], errors="ignore")
nba_recreated_clean = nba_recreated_clean.rename(columns={"Year_x": "Year", "Player_x": "Player"})

display(nba_recreated_clean.head())
display(nba_recreated_clean.tail())
```

#### 10. Subset the original dataset to the years 1995-1997. Group the data by team name and year then summarize the average number of points per team. Arrange from most to least points. <a class="anchor" id="bullet10"></a>

```python
q10 = (
    nba.loc[nba["Year"].between(1995, 1997), ["Tm", "Year", "PTS"]]
       .dropna(subset=["Tm", "Year", "PTS"])
       .groupby(["Tm", "Year"], as_index=False)
       .agg(avg_points=("PTS", "mean"))
       .sort_values("avg_points", ascending=False)
       .reset_index(drop=True)
)

q10["Year"] = q10["Year"].astype(int)

q10.shape
display(q10.head())
display(q10.tail())
q10
```

#### 11. Similar to Q10, start by subsetting the original dataset to the years 1995-1997. Then create a wide dataset that keeps each year in its own row, but spreads team names to multiple columns with each column delineating points per team in each of the 1995-1997 seasons (Hint: Your final answer should have a shape of 3 x 30).  <a class="anchor" id="bullet11"></a>

```python
q11 = (
    q10.pivot(index="Year", columns="Tm", values="avg_points")
       .reset_index()
)

q11.shape
display(q11.head())
display(q11.tail())
q11
```

#### 12. Now return the data from #11 back into a long (tidy) format by moving Tm, Points and Year into their own columns. <a class="anchor" id="bullet12"></a>

```python
q12 = (
    q11.melt(id_vars="Year", var_name="Tm", value_name="avg_points")
)

q12.shape
display(q12.head())
display(q12.tail())
q12
```
