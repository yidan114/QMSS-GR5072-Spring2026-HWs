# Homework 2 - Data Wrangling with Pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/Seasons_Stats_NBA.csv')
print(df.shape)
df.head()
```

## Q1. Which NBA player scored the most points in a single season? What season/year was this?

```python
idx = df['PTS'].idxmax()
q1 = df.loc[[idx], ['Year', 'Player', 'PTS']]
print(q1)
print(f"\n{df.loc[idx, 'Player']} scored the most points ({df.loc[idx, 'PTS']:.0f}) in {df.loc[idx, 'Year']:.0f}.")
```

## Q2. Which NBA player scored the most points in 1991?

```python
df_1991 = df[df['Year'] == 1991]
idx = df_1991['PTS'].idxmax()
q2 = df.loc[[idx], ['Year', 'Player', 'PTS']]
print(q2)
print(f"\n{df.loc[idx, 'Player']} scored the most points ({df.loc[idx, 'PTS']:.0f}) in 1991.")
```

## Q3. Best free throw percentage (FT%) from year 2000 to most recent year

```python
df_2000 = df[df['Year'] >= 2000].copy()
max_ft = df_2000['FT%'].max()
q3 = df_2000[df_2000['FT%'] == max_ft][['Year', 'Player', 'FT%']].sort_values('Player')
print(q3.head())
print(q3.tail())
print(f"\nShape: {q3.shape}")
```

## Q4. Best FT% from 2000 onward among those with at least 50 FTA

```python
df_2000_50 = df_2000[df_2000['FTA'] >= 50].copy()
max_ft_50 = df_2000_50['FT%'].max()
q4 = df_2000_50[df_2000_50['FT%'] == max_ft_50][['Year', 'Player', 'FT%', 'FTA']].sort_values('Player')
print(q4.head())
print(q4.tail())
```

## Q5. Rename the variable "Pos" to "Position"

```python
df = df.rename(columns={'Pos': 'Position'})
print(df[['Player', 'Position']].head())
print(df[['Player', 'Position']].tail())
```

## Q6. Create "first_position" and "second_position" from Position

```python
split_pos = df['Position'].str.split('-', expand=True)
df['first_position'] = split_pos[0]
df['second_position'] = split_pos[1] if 1 in split_pos.columns else np.nan
print(df[['Player', 'Position', 'first_position', 'second_position']].head())
print(df[['Player', 'Position', 'first_position', 'second_position']].tail())
```

## Q7. Combine back into "position_united" and present frequency table

```python
df['position_united'] = df.apply(
    lambda row: row['first_position'] + '-' + row['second_position'] 
    if pd.notna(row['second_position']) else row['first_position'], axis=1
)
print(df[['Player', 'position_united']].head())
print(df[['Player', 'position_united']].tail())
print("\nFrequency table:")
print(df['position_united'].value_counts())
```

## Q8. Create two new datasets and add mergeid

```python
# Reload original data for clean state
df_orig = pd.read_csv('../Data/Seasons_Stats_NBA.csv')

# Dataset 1: all columns except Age
df_no_age = df_orig.drop(columns=['Age']).copy()
df_no_age['mergeid'] = range(1, len(df_no_age) + 1)

# Dataset 2: Year, Player, Age
df_age = df_orig[['Year', 'Player', 'Age']].copy()
df_age['mergeid'] = range(1, len(df_age) + 1)

print("Dataset without Age:")
print(df_no_age.head())
print(df_no_age.tail())
print(f"Shape: {df_no_age.shape}")

print("\nDataset with Year, Player, Age:")
print(df_age.head())
print(df_age.tail())
print(f"Shape: {df_age.shape}")
```

## Q9. Join the two datasets to recreate the original dataset plus mergeid

```python
df_merged = pd.merge(df_no_age, df_age, on='mergeid')
print(df_merged.head())
print(df_merged.tail())
print(f"Shape: {df_merged.shape}")
```

## Q10. Average points per team for 1995-1997, arranged most to least

```python
df_95_97 = df_orig[(df_orig['Year'] >= 1995) & (df_orig['Year'] <= 1997)].copy()
q10 = df_95_97.groupby(['Tm', 'Year'])['PTS'].mean().reset_index()
q10 = q10.sort_values('PTS', ascending=False)
print(q10.head())
print(q10.tail())
print(f"Shape: {q10.shape}")
```

## Q12. Wide dataset: years in rows, team names as columns, values are points (3 x 30)

```python
q12 = q10.pivot(index='Year', columns='Tm', values='PTS')
print(q12.head())
print(q12.tail())
print(f"Shape: {q12.shape}")
```

## Q13. Return Q12 data back to long (tidy) format (90 x 3)

```python
q13 = q12.reset_index().melt(id_vars='Year', var_name='Tm', value_name='PTS')
q13 = q13.sort_values(['Year', 'Tm']).reset_index(drop=True)
print(q13.head())
print(q13.tail())
print(f"Shape: {q13.shape}")
```
